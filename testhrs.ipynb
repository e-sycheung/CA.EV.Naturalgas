```python
import pandas as pd
import numpy as np
import datetime 
```

```python
#Create Filepath and read csv
filepath = r"C:/Users/Elissa/Desktop/CAISO-ALL-SUPPLY-DATA1.csv"
```

```python
#Read the csv file 
supply_df = pd.read_csv(filepath)
#print(supply_df.head(-1))

x = supply_df.loc[(supply_df['Date'] == '12/31/2018') & (supply_df['Time'] == '23:55:00')]
print(x.index)

y = supply_df.loc[(supply_df['Date'] == '12/31/2019') & (supply_df['Time'] == '23:55:00')]
print(y.index)

z = supply_df.loc[(supply_df['Date'] == '12/31/2020') & (supply_df['Time'] == '23:55:00')]
print(z.index)

supply_df['Date'] = pd.to_datetime(supply_df['Date'])
```

```python

baby_df = supply_df['Time'].str.split(":", expand=True)
print(baby_df)

baby_df = baby_df.rename(columns={baby_df.columns[0]: 'Hours'})
baby_df = baby_df.rename(columns={baby_df.columns[1]: 'Minutes'})
baby_df['Year'] = supply_df['Date'].dt.year
baby_df['Year'] = baby_df['Year'].astype(int)
baby_df['Hours'] = baby_df['Hours'].astype(int)
baby_df['Minutes'] = baby_df['Minutes'].astype(int)
baby_df= baby_df.drop([2],1)

baby_df.head(100
)
```

```python
baby_df['DemandMW'] = 0.0

mask = (baby_df['Year'] == 2018) & (baby_df['Hours']<=6) | (baby_df['Hours']>=22)
mask1 = (baby_df['Year'] == 2019) & (baby_df['Hours']<=6) | (baby_df['Hours']>=22)
mask2 = (baby_df['Year'] == 2020) & (baby_df['Hours']<=6) | (baby_df['Hours']>=22)


baby_df.loc[mask,'DemandMW'] = 6.2356
baby_df.loc[mask1,'DemandMW'] = 8.2818
baby_df.loc[mask2,'DemandMW'] = 9.9174
```
