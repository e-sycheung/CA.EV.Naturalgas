```python
# Import the dependencies
import os
import requests
import pandas as pd
```

```python
# 3/16/2022 supply data CSV from CAISO Network XHR
url = 'https://www.caiso.com/outlook/SP/History/20220316/fuelsource.csv?_=1647651289493'
```

```python
# Save CSV to local machine
proxies = {}
response = requests.get(url=url, proxies=proxies)
with open('supply_2022_03_16.csv', 'wb') as f:
    f.write(response.content)
```

```python
# Save CSV to DataFrame
supply_2022_03_16_df = pd.read_csv(url)
supply_2022_03_16_df.head()
```
