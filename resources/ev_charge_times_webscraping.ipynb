```python
# Import the dependencies
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# Get the HTML
url = 'https://suncountryhighway.ca/ev-charge-times/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Get the table
table = soup.find('table', {'id':'tablepress-2'})
```

```python
# Get table headers
headers = []
for i in table.find_all('th'):
    title = i.text
    headers.append(title)
    
headers
```

```python
# Create first table DataFrame
df = pd.DataFrame(columns = headers)

df
```

```python
# Get first table row data
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```

```python
# View DataFrame
df
```

```python
# Check data types
df.dtypes
```

```python
# Correct column types
df['Vehicle'] = df['Vehicle'].astype('str')
df['Acceptance Rate (kW)'] = df['Acceptance Rate (kW)'].astype('float')
df['Battery Size (kWh)'] = df['Battery Size (kWh)'].astype('float')
df['Level 1 PCS-15 1.4kW'] = df['Level 1 PCS-15 1.4kW'].astype('float')
df['Level 2 SCH-15 1.4kW'] = df['Level 2 SCH-15 1.4kW'].astype('float')
df['Level 2 SCH-20 3.8kW'] = df['Level 2 SCH-20 3.8kW'].astype('float')
df['Level 2 SCH-25 4.8kW'] = df['Level 2 SCH-25 4.8kW'].astype('float')
df['Level 2 SCH-30 5.8kW'] = df['Level 2 SCH-30 5.8kW'].astype('float')
df['Level 2 EV-40 7.7kW'] = df['Level 2 EV-40 7.7kW'].astype('float')
df['Level 2 EV-50 9.6kW'] = df['Level 2 EV-50 9.6kW'].astype('float')
df['Level 2 EV-60 11.5kW'] = df['Level 2 EV-60 11.5kW'].astype('float')
```

```python
# Check data types again
df.dtypes
```

```python
# Create the output file (CSV)
output_data_file = "Resources/electric_vehicle_charge_times.csv"

# Export the DataFrame into a CSV without the index
df.to_csv(output_data_file, index=False)
```

```python
# Create new DatbaFrame for MW convervions
df_MW = df
```

```python
# Convert column names and values to MW
df_MW = df_MW.rename({'Acceptance Rate (kW)':'Acceptance Rate (MW)', 'Battery Size (kWh)':'Battery Size (MWh)', 'Level 1 PCS-15 1.4kW':'Level 1 PCS-15 0.0014MW', 'Level 2 SCH-20 3.8kW':'Level 2 SCH-20 0.0038MW', 'Level 2 SCH-25 4.8kW':'Level 2 SCH-25 0.0048MW', 'Level 2 SCH-30 5.8kW':'Level 2 SCH-30 0.0058MW', 'Level 2 EV-40 7.7kW':'Level 2 EV-40 0.0077MW', 'Level 2 EV-50 9.6kW':'Level 2 EV-50 0.0096MW', 'Level 2 EV-60 11.5kW':'Level 2 EV-60 0.00115MW'}, axis='columns')
df_MW
```

```python
# Create the MW output file (CSV)
MW_output_data_file = "electric_vehicle_charge_times_MW.csv"

# Export the MW DataFrame into a CSV without the index
df_MW.to_csv(MW_output_data_file, index=False)
```
